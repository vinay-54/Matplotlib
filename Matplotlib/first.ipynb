```python
# Subplots
# 1. it is used to create multiple plots in a single figure
# 2. it is used to visualize multiple plots together for better comparison
# 3. it is used to create a grid of plots for better visualization
fig,axs=plt.subplots(2, 2, figsize=(8, 6))  # Creating a grid of 2x2 subplots with a figure size of 10x8 inches

axs[0, 0].plot(batsman['index'], batsman['V Kohli'], color="#FF008C", linestyle='--', marker='o', markersize=5)
axs[0, 0].set_title("V Kohli Career")  # Setting title for the first subplot

axs[0, 1].plot(batsman['index'], batsman['RG Sharma'], color="#006AFF", linestyle='--', marker='d', markersize=5)
axs[0, 1].set_title("Rohit Sharma Career")  # Setting title for the second subplot

axs[1, 0].bar(np.arange(df.shape[0]),df['2015'],width=0.4,color='orange')
axs[1, 0].set_title("Batsman Performance in 2015")  # Setting title for the third subplot

axs[1, 1].bar(np.arange(df.shape[0]),df['2016'],width=0.4,color='purple')
axs[1, 1].set_title("Batsman Performance in 2016")  # Setting title for the fourth subplot

for ax in axs.flat:
    ax.set(xlabel='Season', ylabel='Runs Scored')  # Setting x and y labels for all subplots    
    ax.label_outer()  # Hiding x labels and tick labels for top plots and y ticks for right plots
plt.tight_layout()  # Adjusting the layout of the subplots for better visibility 
```

```python
batters_data
# Scatter Plot
fig=plt.figure() # Creating a new figure for the 3D plot
ax=fig.add_subplot(111,projection='3d') # Adding a 3D subplot to the figure
ax.scatter(batters_data['runs'],batters_data['avg'],batters_data['strike_rate'],cmap='jet',s=50) # Creating a 3D scatter plot with avg, strike_rate and fours as the three axes
ax.set_title('Batsman Performance in IPL') # Adding title to the 3D plot
ax.set_xlabel('Runs') # Setting label for the x-axis
ax.set_ylabel('Avg') # Setting label for the y-axis
ax.set_zlabel('Strike Rate') # Setting label for the z-axis
plt.show()  # Show the 3D plot
```

```python
# Line Plot
fig=plt.figure() # Creating a new figure for the 3D plot
x=[0,1,5]
y=[0,2,6]
z=[0,3,7]
ax=fig.add_subplot(111,projection='3d') # Adding a 3D subplot to the figure
ax.plot(x,y,z,marker='o',color="green") # Creating a 3D line plot with x, y and z as the three axes
ax.set_title('3D Line Plot') # Adding title to the 3D plot
```

```python
# 3D Sufrace Plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
np.meshgrid(x,y)
xx,yy=np.meshgrid(x,y)
z=xx**2 + yy**2
print(z.shape)
```

```python
fig1=plt.figure(figsize=(8,6))  # Creating a new figure for the 3D plot with a specific size
ax1=fig1.add_subplot(111,projection='3d') # Adding a 3D subplot to the figure
ax1.plot_surface(xx,yy,z,cmap='viridis') # Creating a 3D surface plot with xx, yy and z as the three axes
ax1.set_title('3D Surface Plot') # Adding title to the 3D plot
```

```python
z2=np.sin(np.sqrt(xx**2 + yy**2))
fig2=plt.figure(figsize=(8,6))  # Creating a new figure for the 3D plot with a specific size
ax2=fig2.add_subplot(111,projection='3d') # Adding a 3D subplot to the figure
ax2.plot_surface(xx,yy,z2,cmap='viridis') # Creating a 3D surface plot with xx, yy and z as the three axes
ax2.set_title('3D Surface Plot') # Adding title to the 3D plot
```

```python
# Functions
z1 = np.sin(np.sqrt(xx**2 + yy**2))
z2 = np.sin(np.sqrt(xx**2 + yy**2))

# Create a figure with 2 subplots (1 row, 2 columns)
fig = plt.figure(figsize=(12, 5))

# Left: Contour Plot
ax1 = fig.add_subplot(1, 2, 2)
contour = ax1.contourf(xx, yy, z1, cmap='viridis')
ax1.set_title('Contour Plot')
fig.colorbar(contour, ax=ax1)  # Add colorbar

# Right: 3D Surface Plot
ax2 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax2.plot_surface(xx, yy, z2, cmap='viridis')
ax2.set_title('3D Surface Plot')
fig.colorbar(surf, ax=ax2, shrink=0.5)  # Add colorbar

plt.tight_layout()
plt.show()
```

```python
# One figure with 2 subplots side by side
fig = plt.figure(figsize=(12, 5))

# Left: 3D Surface Plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(xx, yy, z, cmap='viridis')
ax1.set_title('3D Surface Plot')
fig.colorbar(surf, ax=ax1, shrink=0.5)

# Right: 2D Contour Plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contour(xx, yy, z, cmap='viridis')
ax2.set_title('2D Contour Plot')
fig.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```
